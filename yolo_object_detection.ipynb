```python
from ultralytics import YOLO
```

```python
model = YOLO("yolo11x.pt")
```

```python
model.info()
```

```python
import cv2 as cv
cap = cv.VideoCapture(0)

while cap.isOpened():
    ret, frame = cap.read()
    cv. imshow("Yolo", frame)

    if cv.waitKey(1) & 0xFF == ord("q"):
        break
cap.release()
cv.destroyAllWindows()
```

```python
results = model(source=0, show=True)
```
