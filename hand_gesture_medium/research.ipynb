```python
import cv2 as cv
```

```python
img = cv.imread("D:/mediapipe_projects/hand_gesture/data/ronaldo.jpg")
```

```python
# cv.imshow("Picture", img)
```

```python
import matplotlib.pyplot as plt  
import matplotlib.image as mpimg  

# Load the image  
# img = mpimg.imread(img)  
img1 = cv.cvtColor(img, cv.COLOR_RGB2BGR)
# Create a figure and axis  
fig, ax = plt.subplots()  

# Display the image  
ax.imshow(img1)  

# Hide the axes  
# ax.axis('off')  

# Add text above the image  
fig.text(0.5, 0.9, 'RGB to BGR - Original', fontsize=14, color='red', ha='center')  

# Show the plot  
plt.show()  
```

```python
import matplotlib.pyplot as plt  
import matplotlib.image as mpimg  

# Load the image  
# img = mpimg.imread(img)  
img2 = cv.cvtColor(img1, cv.COLOR_BGR2RGB)
# Create a figure and axis  
fig, ax = plt.subplots()  

# Display the image  
ax.imshow(img2)  

# Hide the axes  
# ax.axis('off')  

# Add text above the image  
fig.text(0.5, 0.9, 'BGR To RGB', fontsize=14, color='red', ha='center')  

# Show the plot  
plt.show()  
```
